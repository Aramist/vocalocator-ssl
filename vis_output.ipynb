```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fpath = 'sample_output.npy'
inpath = 'sample_input.npy'
```

```python
samp_out = np.load(fpath)
samp_in = np.load(inpath)
```

```python
def vis(index):
    plt.figure(0, figsize=(10, 10))
    plt.title('Output')
    # each sample is four channel audio with shape (8192, 4)
    for n in range(4):
        plt.subplot(221 + n)
        plt.plot(samp_out[index, :, n])
        
    plt.figure(1, figsize=(10, 10))
    plt.title('Output')
    # each sample is four channel audio with shape (8192, 4)
    for n in range(4):
        plt.subplot(221 + n)
        plt.specgram(samp_out[index, :, n], Fs=125000)
        
    plt.figure(2, figsize=(10, 10))
    plt.title('Input')
    # each sample is four channel audio with shape (8192, 4)
    for n in range(4):
        plt.subplot(221 + n)
        plt.specgram(samp_in[index, :, n], Fs=125000)
    plt.show()
```

```python
vis(2)
```
